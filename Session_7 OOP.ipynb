#### ATM Machine Application with OOP

```python
class Atm:
    # Constructor(special Function)-> superpower -> we dont need to call the constructor by ourself in code for execution
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input("""
        Hi How can i help you
        1.Press 1 to create pin
        2.Press 2 to change pin
        3.Press 3 to check balance
        4.Press 4 to withdraw
        5.Anything Else to exit
        """)
    
        if user_input == '1':
            #create pin
            self.create_pin()
        elif user_input == '2':
            #change pin
            self.change_pin()
        elif user_input == '3':
            # check Balance
            self.check_balance()
        elif user_input == '4':
            # withdraw
            self.withdraw()
        else:
            exit()

    # Create the pin
    def create_pin(self):
        user_pin = input("Enter your pin")
        self.pin = user_pin

        user_balance = int(input("Enter your balance"))
        self.balance = user_balance

        print("Pin Crated Succesfully")
        self.menu()

    # Change the pin
    def change_pin(self):
        old_pin = input("Enter old pin")

        if old_pin == self.pin:
            #let him change the pin
            new_pin = input("Enter New Pin")
            self.pin = new_pin
            print("Pin Changed Successfully")
            self.menu()
        else:
            print("Not Allowed")
            self.menu()

    #Check the balance
    def check_balance(self):
        user_pin = input("Enter your pin")
        if user_pin == self.pin:
            print("Your Available Balance is :",self.balance)
            self.menu()
        else:
            print("Kindly Enter the valid Pin")
            self.menu()

    # Withdraw
    def withdraw(self):
        user_pin = input("Enter your pin")
        if user_pin == self.pin:
            user_amount = int(input("Enter Amount You Want To Withdraw"))
            if user_amount <= self.balance:
                self.balance = self.balance - user_amount
                print("Take Your Amount")
                print("Your Remaining Balance is :",self.balance)
            else:
                print("Check Your Balance We are not able to proceed your request")
        else:
            print("Kindly Check Your Pin")
        self.menu()
        
        
        
        
```

## Creation of Datatype 

### 1] Fractions

```python
class fraction:
    #Parameterized constructor(needsinput)
    def __init__(self,x,y):                    # First Magic Method
        self.num = x
        self.den = y

    def __str__(self):                         # Second Magic Method     
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):                  # Third Magic Method
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,other):                  # Fourth Magic Method
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __mul__(self,other):                  # Fifth Magic Method
        new_num = self.num*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __truediv__(self,other):                  # Sixth Magic Method
        new_num = self.num/other.den
        new_den = self.den*other.num
        return '{}/{}'.format(new_num,new_den)

    def convert_to_decimal(self):
        return self.num/self.den

    
    
```

```python
fr1 = fraction(3,4)
fr2 = fraction(1,2)
```
